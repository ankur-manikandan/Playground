```python
%matplotlib notebook

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

SEED = 42
np.random.seed(SEED)
```

```python
log_vals = np.log(range(50000))

fig, ax = plt.subplots()
plt.plot(log_vals)
plt.show()
```

```python
x = range(-100, 10)
exp_vals = np.exp(x)

fig, ax = plt.subplots()
plt.plot(x, exp_vals)
plt.show()
```
