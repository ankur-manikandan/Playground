```python
%matplotlib notebook

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ts = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
```

```python
fig, ax = plt.subplots()
ts.plot(ax=ax)
plt.show()
```

```python
df3 = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
```

```python
# fig, ax = plt.subplots(nrows=2, ncols=1)
# df3.plot.scatter(x='a', y='b')
# plt.show()

fig, ax = plt.subplots()
df3.groupby('class')['a', 'b'].count().plot.bar(ax=ax, subplots=True)
plt.show()
```

```python
fig, ax = plt.subplots()
df3.plot.scatter(x='a', y='b', color='blue', label='points A', ax=ax)
df3.plot.scatter(x='c', y='d', color='green', label='points B', ax=ax)
plt.margins(0.1)
plt.show()
```

```python
df3.head()
```

```python
df3['class'] = np.random.choice(['y', 'n'], p=[0.2, 0.8], size=len(df3), replace=True)
df3.head()
```

```python
fig, ax = plt.subplots()
df3.groupby('class')['a', 'b'].count().plot.bar(ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots()
df3.loc[(df3['class'] == 'y'), 'a'].hist(label='yes', ax=ax)
df3.loc[(df3['class'] == 'n'), 'b'].hist(label='no', ax=ax, alpha=0.3)
plt.show()
```

```python
fig, ax = plt.subplots()
df3.loc[df3['class'] == 'y'].plot.scatter(x='a', y='b', color='green', label='yes', ax=ax)
df3.loc[df3['class'] == 'n'].plot.scatter(x='a', y='b', color='red', label='no', ax=ax)
plt.show()
```
